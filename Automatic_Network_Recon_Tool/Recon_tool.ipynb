```python
import subprocess

target = input("Enter target domain or IP: ")

result = subprocess.run(["ping", target], capture_output=True, text=True)

print(result.stdout)
```

```python
import subprocess

target = input("Enter target domain or IP: ")

print("\nRunning ping...\n")
ping = subprocess.run(["ping", target], capture_output=True, text=True)
print(ping.stdout)

print("\nRunning nslookup...\n")
nslookup = subprocess.run(["nslookup", target], capture_output=True, text=True)
print(nslookup.stdout)
```

```python
print("\nRunning tracert...\n")
traceroute = subprocess.run(["tracert", target], capture_output=True, text=True)
print(traceroute.stdout)
```

```python
import subprocess

target = input("Enter target domain or IP: ")

with open("recon_report.txt", "w") as report:

    report.write("=== RECON REPORT ===\n\n")

    report.write("PING RESULTS:\n")
    ping = subprocess.run(["ping", target], capture_output=True, text=True)
    report.write(ping.stdout)

    report.write("\nNSLOOKUP RESULTS:\n")
    nslookup = subprocess.run(["nslookup", target], capture_output=True, text=True)
    report.write(nslookup.stdout)

    report.write("\nTRACERT RESULTS:\n")
    traceroute = subprocess.run(["tracert", target], capture_output=True, text=True)
    report.write(traceroute.stdout)

print("Recon complete. Results saved to recon_report.txt")
```
