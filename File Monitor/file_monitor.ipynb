```python
import os
import hashlib

folder_path = r"C:\Users\Sanjo\Desktop\Py Auto\monitor_folder"


file_hashes = {}

for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)

    if os.path.isfile(file_path):
        with open(file_path, "rb") as f:
            file_data = f.read()
            file_hash = hashlib.sha256(file_data).hexdigest()
            file_hashes[filename] = file_hash

print("Baseline hashes created:")
for file, hash_value in file_hashes.items():
    print(file, ":", hash_value)
```

```python
with open("baseline.txt", "w") as baseline:
    for file, hash_value in file_hashes.items():
        baseline.write(f"{file}:{hash_value}\n")

print("Baseline saved to baseline.txt")
```

```python
import os
import hashlib

folder_path = r"C:\Users\Sanjo\Desktop\Py Auto\monitor_folder"

# Load baseline
baseline_hashes = {}

with open("baseline.txt", "r") as baseline:
    for line in baseline:
        file, hash_value = line.strip().split(":")
        baseline_hashes[file] = hash_value

# Check current files
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)

    if os.path.isfile(file_path):
        with open(file_path, "rb") as f:
            file_data = f.read()
            current_hash = hashlib.sha256(file_data).hexdigest()

        if filename not in baseline_hashes:
            print(f"New file detected: {filename}")

        elif baseline_hashes[filename] != current_hash:
            print(f"File modified: {filename}")

# Check for deleted files
for filename in baseline_hashes:
    if filename not in os.listdir(folder_path):
        print(f"File deleted: {filename}")
```
