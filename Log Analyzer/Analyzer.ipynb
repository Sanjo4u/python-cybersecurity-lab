```python
import re

file = open("C:\\Users\\Sanjo\\Desktop\\Py Auto\\log analyzer\\logs\\auth.logs.txt", "r")


lines = file.readlines()
file.close()

failed_count = 0
ip_addresses = set()

for line in lines:
    if "failed" in line.lower():
        failed_count += 1

    ips = re.findall(r"\d+\.\d+\.\d+\.\d+", line)
    for ip in ips:
        ip_addresses.add(ip)

print("Failed logins:", failed_count)
print("\nSuspicious IPs:")
for ip in ip_addresses:
    print(ip)
```
